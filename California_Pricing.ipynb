```python
# standard imports
import pandas as pd
import numpy as np

# Do not change this option; This allows the CodeGrade auto grading to function correctly
pd.set_option('display.max_columns', 20)
```

```python
## Viewing  long_lat CSV file
long_lat1 = pd.read_csv('long_lat.csv')
long_lat1
```

```python
## Viewing cal_housing_low CSV file
cal_housing_low1 = pd.read_csv('cal_housing_low.csv')
cal_housing_low1
```

```python
## Viewing cal_housing_medium CSV file
cal_housing_medium1 = pd.read_csv('cal_housing_medium.csv')
cal_housing_medium1
```

```python
## Viewing cal_housing_high CSV file
cal_housing_high1 = pd.read_csv('cal_housing_high.csv')
cal_housing_high1
```

```python
## Viewing ocean_proximity CSV file
ocean_proximity1 = pd.read_csv('ocean_proximity.csv')
ocean_proximity1
```

```python
# Using .merge() and .concat() to merge the 5 CSV files
housing1 = pd.merge(long_lat1, ocean_proximity1, on = ('longitude', 'latitude'), how = 'inner')
housing2 = pd.concat([cal_housing_low1, cal_housing_medium1, cal_housing_high1])
housing = pd.merge(housing2, housing1, on = 'id')
housing
```

```python
# Sorting the housing data frame by id in ascending order
housing = housing.sort_values(by = 'id')
housing
```

```python
## Droping any missing values from the medianHouseValue column
housing = housing.dropna(subset = ['medianHouseValue'])
housing
```

```python
## Filling missing values with the median in the column housingMedianAge with the housing median age that we will calculate
median_value = housing['housingMedianAge'].median()
housing['housingMedianAge'] = housing['housingMedianAge'].fillna(median_value)
housing
```

```python
## Dropping the state column
housing = housing.drop(columns = ['state'])
housing
```

```python
## Scaling the medianIncome in $10,000 of dollars
housing['medianIncome'] = housing['medianIncome'].div(10000)
housing
```

```python
## Changing any values in medianIncome under 0.4999 to 0.4999 and any vales above 15.0001 to 15.0001
housing['medianIncome'] = np.clip(housing['medianIncome'], 0.4999, 15.0001)
housing
```

```python
## Reverting the medianHouseValue back to actual dollars
housing['medianHouseValue'] = housing['medianHouseValue'].mul(1000)
housing
```

```python
## Changing column names
housing = housing.rename(columns = {'housingMedianAge': 'housing_median_age', 'totalRooms' : 'total_rooms', 'totalBedrooms' : 'total_bedrooms', 'medianIncome' : 'median_income', 'medianHouseValue' : 'median_house_value' })
housing
```

```python
## Changing the order of the columns
housing = housing[['longitude', 'latitude', 'housing_median_age', 'total_rooms', 'total_bedrooms', 'population', 'households', 'median_income', 'median_house_value', 'ocean_proximity']]
housing
```

```python
## Change all columns except ocean_proximity to a float
ocean_proximity_no_change = 'ocean_proximity'
def convert_to_float(col):
    if col.name != ocean_proximity_no_change:
        return col.astype(float)
    else:
        return col
housing = housing.apply(convert_to_float)
housing
```

```python
## Resetting the data frame's index from 0 to n-1 using drop = True to exclude index column 
housing = housing.reset_index(drop = True)
housing
```
